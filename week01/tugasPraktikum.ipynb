<b>1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan. 

<b>Jawab:</b>

Variable yang dapat digunakan menurut saya adalah 'diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean' 

```python
import pandas as pd

df = pd.read_csv('data/wbc.csv')
df.head()
```

```python
df = df.dropna(axis = 1)
df.head()
```

```python
df = df[['diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean']]
df.head()
```

<b>2. Lakukan proses encoding pada kolom "diagnosis". 

<b>Jawab:</b>

Proses encoding pada diagnosis dilakukan sebagai langkah yang perlu dilakukan karena algoritma pembelajaran mesin melakukan proses kalkulasi secara matematis.

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

le = LabelEncoder() # membuat objek dari LabelEncoder
df['diagnosis'] = le.fit_transform(df['diagnosis']) # proses encoding
```

```python
df.head()
```

<b>3. Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik. 

<b>Jawab:</b>

Standarisasi adalah proses yang penting dalam pengolahan data dan pembelajaran mesin yang bertujuan untuk mengubah data sehingga memiliki mean (rata-rata) nol dan deviasi standar (standar deviasi) satu. 

```python
numeric_columns = df.columns.difference(['diagnosis'])
df[numeric_columns]
```

<b>4. Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

<b>Jawab:</b>

Stratified splitting data adalah metode pembagian data yang mempertahankan proporsi kelas atau kelompok yang sama dalam setiap subset data yang dihasilkan.

```python
df2 = pd.read_csv('data/wbc.csv')

# Split data
from sklearn.model_selection import train_test_split

# Rasio yang akan kita gunakan adalah 8:2
df2_train, df2_test = train_test_split(df2, test_size=0.2, random_state=0, stratify=df['diagnosis'])

# Cek masing-masing ukuran data
print(f'Jumlah label data train:\n{df2_train.diagnosis.value_counts()}')
print(f'Jumlah label data test:\n{df2_test.diagnosis.value_counts()}')
```
